```python
# FOOD DELIVERY APP - SINGLE CELL (JUPYTER)

from flask import Flask, render_template_string, request, jsonify

app = Flask(__name__)

foods = [
    {"id":1, "name":"Pizza", "price":250},
    {"id":2, "name":"Burger", "price":120},
    {"id":3, "name":"Pasta", "price":180},
    {"id":4, "name":"Biryani", "price":220}
]

cart = []

HTML = """
<!DOCTYPE html>
<html>
<head>
<title>Food Delivery App</title>
<style>
body{font-family:Arial;background:#f4f4f4;padding:20px}
.dark{background:#121212;color:white}
.food{border:1px solid #ccc;padding:10px;margin:10px}
button{padding:5px 10px}
#cart{margin-top:20px}
</style>
</head>

<body id="body">
<h2>üçî Food Delivery App</h2>
<button onclick="toggle()">üåô Dark Mode</button>

<div id="foods"></div>

<h3>üõí Cart</h3>
<ul id="cart"></ul>

<script>
let dark=false;

function toggle(){
  dark=!dark;
  document.getElementById("body").className = dark?"dark":"";
}

fetch("/foods")
.then(res=>res.json())
.then(data=>{
  let f="";
  data.forEach(i=>{
    f+=`<div class='food'>
         <b>${i.name}</b> - ‚Çπ${i.price}
         <button onclick="add(${i.id})">Add</button>
         </div>`;
  })
  document.getElementById("foods").innerHTML=f;
})

function add(id){
 fetch("/add/"+id)
 .then(res=>res.json())
 .then(data=>{
   let c="";
   data.forEach(i=>c+=`<li>${i.name} - ‚Çπ${i.price}</li>`)
   document.getElementById("cart").innerHTML=c;
 })
}
</script>
</body>
</html>
"""

@app.route("/")
def home():
    return render_template_string(HTML)

@app.route("/foods")
def get_foods():
    return jsonify(foods)

@app.route("/add/<int:id>")
def add_cart(id):
    for f in foods:
        if f["id"]==id:
            cart.append(f)
    return jsonify(cart)

app.run()
```
