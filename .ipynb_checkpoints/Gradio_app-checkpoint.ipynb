```python
import gradio as gr
```

```python
# jupyter refresh issue fix
gr.close_all()
```

```python
def add_numbers(a, b):
    return a + b

interface = gr.Interface(
    fn=add_numbers,
    inputs=[
        gr.Number(label="First Number"),
        gr.Number(label="Second Number")
    ],
    outputs=gr.Number(label="Sum"),
    title="Addition Calculator",
    description="Enter two numbers and get their sum."
)
```

```python
interface.launch(share=False)   # share=True bhi kar sakte ho
```
