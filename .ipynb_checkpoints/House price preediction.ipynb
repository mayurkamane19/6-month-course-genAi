```python
import streamlit as st
import pandas as pd
from sklearn.linear_model import LinearRegression

# Title
st.title("üè† House Price Prediction App")

# Sample dataset
data = {
    "Area": [1000, 1500, 2000, 2500, 3000],
    "Bedrooms": [2, 3, 3, 4, 4],
    "Bathrooms": [1, 2, 2, 3, 3],
    "Price": [50, 65, 80, 100, 120]
}

df = pd.DataFrame(data)

# Train model
X = df[["Area", "Bedrooms", "Bathrooms"]]
y = df["Price"]

model = LinearRegression()
model.fit(X, y)

st.subheader("üìä Training Data")
st.dataframe(df)

# User input
st.subheader("üßÆ Enter House Details")

area = st.number_input("Area (sq ft)", 500, 5000, step=100)
bedrooms = st.number_input("Bedrooms", 1, 10)
bathrooms = st.number_input("Bathrooms", 1, 10)

# Prediction
if st.button("Predict Price"):
    prediction = model.predict([[area, bedrooms, bathrooms]])
    st.success(f"üè∑Ô∏è Estimated House Price: ‚Çπ {prediction[0]:.2f} Lakhs")
```
