```python
class computer():        
  
    def config(self):     
        print('i5', '8gb')
    def config2(self):
        print('i7', '32gb')
    def __init__(self):
        print('in init')

com1 = computer()
com2 = computer()

com1.config() 
com1.config2()
```

```python
class computer():        
    
    def __init__(self): #__init__ means to initialize the variable
        print('in init')
    
    def config(self):     
        print('i5', '8gb')

com1 = computer('i5', 8)

com1.config()
```

```python
class computer():        
    
    def __init__(self, cpu, ram, price): # cpu & ram as argument
        print('in init')
    
    def config(self):     
        print('i5', '8gb', 50000)
        print('i7', 16, 100000)

com1 = computer('i5', 8, 50000)
com2 = computer('i7', 16, 10000)

com1.config() 
com2.config()
```

```python
class computer():        
    
    def __init__(self,cpu,ram): 
        self.cpu = cpu 
        self.ram = ram 

    def config(self):     
        print('config is', self.cpu, self.ram)

com1 = computer('i5', 8)
com2 = computer('i7', 16)


com1.config() 
com2.config()
```

```python
class computer():        
    
    def __init__(self,cpu,ram, price): 
        self.cpu = cpu 
        self.ram = ram 
        self.price = price 

    def config(self):     
        print('config is', self.cpu, self.ram, self.price )

com1 = computer('i5', 8, 50000)
com2 = computer('i7', 16, 70000)


com1.config() 
com2.config()
```

```python
class computer():
    def __init__(self):
        self.name = 'mayur'
        self.age = 37

c1 = computer()
c2 = computer()

print(c1.name, c1.age)
print(c2.name)
```

```python
class computer():
    def __init__(self):
        self.name = 'mayur'
        self.age = 37

c1 = computer()
c2 = computer()

c2.name = 'mit'
c1.name = 'mayur'

print(c1.name)
print(c2.name)
```
