```python
class student():
    def __init__(self,m1,m2):
        self.m1 = m1
        self.m2 = m2
        
    def __add__(self, other): ## WE DEFINE THE ADD METHOD . THIS CONCEPT IS CALLED OPERATOR OVERLOADING
        m1 = self.m1 + other.m1
        m2 = self.m2 + other.m2
        s3 = student(m1,m2)
        
        return s3   
 
s1 = student(60,70)
s2 = student(80,90)

s3 = s1 + s2 
# when we create add method in above behind the memory s1 & s2 will convert to self & other respectivly
print(s3.m1)
print(s3.m2)

# if you want to add 2 student then operator need to overload 
```

```python
#if the see the above error we got error as > is not support so we need to define the > method (gt)
class student():
    
    def __init__(self,m1,m2):
        self.m1 = m1
        self.m2 = m2
        
    def __add__(self, other): ## WE DEFINE THE ADD METHOD . THIS CONCEPT IS CALLED OPERATOR OVERLOADING
        m1 = self.m1 + other.m1
        m2 = self.m2 + other.m2
        s3 = student(m1,m2)
        
        return s3  
    def __gt__(self,other):
        r1 = self.m1 + self.m2
        r2 = other.m1 + other.m2
        
        if r1 > r2:
            return True
        else:
            return False
 
s1 = student(70,50)
s2 = student(80,90)

s3 = s1 + s2 

if s1>s2:
    print('s1 wins')
else:
    print('s2 wins')
    
# we are not define it so how to check it 
```

```python
class student():
    def __init__(self,m1,m2):
        self.m1 = m1
        self.m2 = m2
        
    def sum(self, a=None, b=None,c=None, d = None, e = None):
        if a!=None and b!=None and c!=None and d!=None and e!=None:
            s = a +b +c + d + e
        elif a!=None and b!=None and c!=None and d!=None :
            s = a +b +c + d
        elif a!=None and b!=None and c!=None:
            s = a + b + c
        elif a!=None and b!=None:
            s = a+b
        elif a!= None and c!= None:
            s = a + c
        else:
            s = a
            
        return s
    
s1 = student(80, 90)

print(s1.sum(10)) # hear we can pass 1 parameter
print(s1.sum(10,20)) # hear we can pass 2 parameter
print(s1.sum(10,20,30)) # hear we can pass 3 parameter
print(s1.sum(10,30))
print(s1.sum(10,20,30,40))
print(s1.sum(10, 20, 30, 40, 50))

# THIS CONCEPT CALLED AS METHOD OVERLOADING
```

Python decorator

```python
def div(a,b):
    print(a / b)
div(4,2)
# but what if we pass the value 2, 4 
```

```python
def div(a,b):
    print(a / b)
div(2,4)
```

```python
def div(a,b):
    if a<b:
        a,b = b,a
    print(a / b)
    
div(2,4)
```

```python
# using help of the decorator you can add the extra feature in the exicting function 

def div(a,b):
    print(a / b)

def div_decorator(func): # hear div_dectorator will accept the div function
    def inner(a,b):
        if a<b:
            a,b = b,a
        return func(a,b)
    return inner

div = div_decorator(div)

div(2,4)     
```

```python
def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper

@my_decorator
def say_hello():
    print("Hello!")

say_hello()
```
