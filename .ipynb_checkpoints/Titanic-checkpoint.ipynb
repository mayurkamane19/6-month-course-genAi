```python
# Remove Cabin column safely
titanic.drop(columns=["Cabin"], inplace=True, errors="ignore")

# Show first 5 rows
titanic.head()
```

```python
def getNumber(str):
    if str=="male":
        return 1
    else:
        return 2
titanic["Gender"]=titanic["Sex"].apply(getNumber)
titanic.head()
```
