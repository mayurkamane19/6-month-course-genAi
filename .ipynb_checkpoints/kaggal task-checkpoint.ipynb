```python
t = movies.merge(tag, on='movieId', how='inner')
t.head()
```

```python
avg_rating= rating.groupby('movieId', as_index=False).mean()
del avg_rating['userId']
avg_rating.head()
```

```python
box_office = movies.merge(avg_rating, on='movieId', how='inner')
box_office.tail()
```

```python
is_highly_rated = box_office['rating'] >= 4.0
box_office[is_highly_rated][-5:]
```

```python
is_Adventure = box_office['genres'].str.contains('Adventure')
box_office[is_Adventure][:5]
```

```python
box_office[is_Adventure & is_highly_rated][-5:]
```
