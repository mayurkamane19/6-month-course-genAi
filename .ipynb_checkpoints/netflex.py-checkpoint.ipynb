```python
# üì¶ Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# üìÅ Load the dataset
df = pd.read_csv(r"C:\Users\mayur\Downloads\archive (3).zip")  # Make sure this file is in the same folder

print('Shape of dataset:', df.shape)
df.head()
```

```python
# Fill missing values and create helper columns
df['country'] = df['country'].fillna('Unknown')
df['rating'] = df['rating'].fillna('Unknown')
df['listed_in'] = df['listed_in'].fillna('')
df['description'] = df['description'].fillna('')

# First country (for simpler grouping)
df['first_country'] = df['country'].apply(lambda x: str(x).split(',')[0].strip())

df.info()
```

```python
# Basic counts and summary
total_titles = len(df)
total_movies = (df['type'] == 'Movie').sum()
total_tvshows = (df['type'] == 'TV Show').sum()
min_year = df['release_year'].min()
max_year = df['release_year'].max()

print('Total titles:', total_titles)
print('Total Movies:', total_movies)
print('Total TV Shows:', total_tvshows)
print('Year range:', min_year, '-', max_year)
```

```python
# Function to search titles by name (partial match)
def search_titles(query, max_results=20):
    mask = df['title'].str.contains(query, case=False, na=False)
    results = df[mask][['title', 'type', 'release_year', 'rating', 'listed_in']].head(max_results)
    return results

# Example search
search_titles('Love')
```

```python
# Create combined text for each title
df['combined_text'] = df['title'] + ' ' + df['listed_in'] + ' ' + df['description']

vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(df['combined_text'])

cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Map title ‚Üí index
indices = pd.Series(df.index, index=df['title'].str.lower()).drop_duplicates()
len(indices)
```

```python
def recommend(title, n=5):
    """Return top N similar titles for a given title."""
    title_lower = title.lower()
    if title_lower not in indices:
        print('Title not found in dataset.')
        return pd.DataFrame()
    idx = indices[title_lower]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:n+1]  # skip itself
    movie_indices = [i[0] for i in sim_scores]
    return df.iloc[movie_indices][['title', 'type', 'release_year', 'rating', 'listed_in', 'description']]

# Try recommendation (change title as per your dataset)
df['title'].sample(5, random_state=42)
```

```python
# Example: replace with any real title from above sample
recommend('3 Idiots', n=5)  # change this name if not present in dataset
```

```python
def filter_titles(type_sel='Any', country_sel='', genre_sel='', year_from=None, year_to=None):
    temp = df.copy()

    if type_sel != 'Any':
        temp = temp[temp['type'] == type_sel]

    if country_sel:
        temp = temp[temp['first_country'].str.lower() == country_sel.lower()]

    if genre_sel:
        temp = temp[temp['listed_in'].str.contains(genre_sel, case=False, na=False)]

    if year_from is not None:
        temp = temp[temp['release_year'] >= year_from]

    if year_to is not None:
        temp = temp[temp['release_year'] <= year_to]

    return temp[['title', 'type', 'release_year', 'first_country', 'rating', 'listed_in']]
    # Example filter
filter_titles(type_sel='Movie', country_sel='India', genre_sel='Drama', year_from=2010, year_to=2021).head(20)
```
