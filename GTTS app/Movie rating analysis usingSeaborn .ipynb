```python
# python is not vectorize programming language
# Building dashboards (dashboard - combination of charts)
sns.set_style('darkgrid')
f, axes = plt.subplots(2, 2, figsize=(15, 15))
# KDE plots
k1 = sns.kdeplot(x=movies.BudgeMillionce, y=movies.AudienceRatings, ax=axes[0,0])
k2 = sns.kdeplot(x=movies.BudgeMillionce, y=movies.CriticRating, ax=axes[0,1])
k1.set(xlim=(-20,160))
k2.set(xlim=(-20,160))
# Violin plot
z = sns.violinplot(
data=movies[movies.Genre == 'Drama'],
x='Year',
y='CriticRating',
ax=axes[1,0]
)
# Scatter style KDE
k4 = sns.kdeplot(
x=movies.CriticRating,
y=movies.AudienceRatings,
fill=True,
ax=axes[1,1],
cmap='Reds'
)
k4b = sns.kdeplot(
x=movies.CriticRating,
y=movies.AudienceRatings,
ax=axes[1,1],
cmap='Reds'
)
plt.show()
```

```python
# How can you style your dashboard using different color map
# python is not vectorize programming language
# Building dashboards (dashboard - combination of charts)
sns.set_style('dark', {'axes.facecolor': 'black'})
f, axes = plt.subplots(2, 2, figsize=(15, 15))
# plot [0,0]
k1 = sns.kdeplot(
x=movies.BudgeMillionce,
y=movies.AudienceRatings,
fill=True,
cmap='inferno',
ax=axes[0, 0]
)
k1b = sns.kdeplot(
x=movies.BudgeMillionce,
y=movies.AudienceRatings,
cmap='cool',
ax=axes[0, 0]
)
# plot [0,1]
k2 = sns.kdeplot(
x=movies.BudgeMillionce,
y=movies.CriticRating,
fill=True,
    cmap='inferno',
    ax=axes[0,1]
)
k2b=sns.kdeplot(
    x=movies.BudgeMillionce,
    y=movies.CriticRating,
    cmap='cool',
    ax=axes[0, 1]
)

# plot [1,0]
z = sns.violinplot(
    data=movies[movies.Genre == 'Drama'],
    x='Year',
    y='CriticRating',
    ax=axes[1,0]
)

# plot [1,1]
k4 = sns.kdeplot(
    x=movies.CriticRating,
    y=movies.AudienceRatings,
    fill=True,
    cmap='Blues_r',
    ax=axes[1,1]
)

k4b = sns.kdeplot(
    x=movies.CriticRating,
    y=movies.AudienceRatings,
    cmap='gist_gray_r',
    ax=axes[1,1]
)

# Limits
k1.set(xlim=(-20, 160))
k2.set(xlim=(-20, 160))

plt.show()
```
