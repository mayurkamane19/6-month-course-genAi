Keyword argument

```python
def person(name,age):
    print(name)
    print(age+1)
    
person(age=23, name='nit')
```

```python
def person(name,age):
    print(name)
    print(age+1)

person(age1=23, name='nit')
```

```python
def person(name,age1):
    print(name)
    print(age1+1)

person(age1=23, name='nit')
```

```python
def person(name,age,city):
    print(name)
    print(age+1)
    print(city)
    
person(age=23, name='nit', city = 'hyd')
```
