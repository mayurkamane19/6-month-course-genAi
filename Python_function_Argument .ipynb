```python
x = 10  # Global variable

def update_x():
    global x  # Declare that we are using the global variable x
    x += 5    # Modify the global variable
   
update_x()
print(x)  # Output: 15
```

```python
# in progammin we need to continue these process thats why we need to use loop hear

def fib(n):
    a = 0
    b = 1
    
    print(a)
    print(b)
    
    for i in range(0, n):
        c = a + b
        a = b
        b = c
        
    print(c)
        
fib(10)
```

```python
def fib(n):
    a, b = 0, 1

    while a <= n:
        print(a)     # 0 se n tak ke Fibonacci numbers print honge
        a, b = b, a + b

fib(89)
```

python decorator

```python
def div(a,b):
    print(a / b)
div(4,2)
```

```python
def div(a, b):
    print(a / b)
div(2,4)
```

```python
def div(a,b):
    
    if a<b:
        a,b = b,a
    print(a / b)
    
div(2,4)
```

```python
def div(a,b):
    print(a / b)

def div_decorator(func): # hear div_dectorator will accept the div function
    def inner(a,b):
        if a<b:
            a,b = b,a
        return func(a,b)
    return inner

div = div_decorator(div)

div(2,4)     
```
