```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# =====================================================
# Load Dataset
# =====================================================
titanic = pd.read_csv(
    r"C:\Users\mayur\Downloads\Titanic-Dataset.csv"
)

print(titanic.tail())
print(titanic.describe())

# =====================================================
# Drop unnecessary columns
# =====================================================
titanic.drop(columns=["Name", "Ticket", "Fare", "Cabin"], inplace=True)

print(titanic.head())

# =====================================================
# Convert Sex to numeric
# male -> 1, female -> 2
# =====================================================
titanic["Gender"] = titanic["Sex"].map({"male": 1, "female": 2})

# Drop old Sex column
titanic.drop(columns=["Sex"], inplace=True)

print(titanic.head())

# =====================================================
# Check missing values
# =====================================================
print(titanic.isnull().sum())

# =====================================================
# Handle Age missing values (Survived = 1)
# =====================================================
meanS = titanic.loc[titanic["Survived"] == 1, "Age"].mean()

titanic["Age"] = np.where(
    titanic["Age"].isnull() & (titanic["Survived"] == 1),
    meanS,
    titanic["Age"]
)

# =====================================================
# Handle Age missing values (Survived = 0)
# =====================================================
meanNS = titanic.loc[titanic["Survived"] == 0, "Age"].mean()
titanic["Age"].fillna(meanNS, inplace=True)

print(titanic.isnull().sum())

# =====================================================
# Embarked survival counts
# =====================================================
for port in ["Q", "C", "S"]:
    survived = titanic.loc[
        (titanic["Embarked"] == port) & (titanic["Survived"] == 1)
    ].shape[0]
    not_survived = titanic.loc[
        (titanic["Embarked"] == port) & (titanic["Survived"] == 0)
    ].shape[0]

    print(f"{port} -> Survived: {survived}, Not Survived: {not_survived}")

# =====================================================
# Convert Embarked to numeric
# S -> 1, Q -> 2, C -> 3
# =====================================================
titanic["Embarked"] = titanic["Embarked"].map({"S": 1, "Q": 2, "C": 3})

# =====================================================
# Rename Gender back to Sex (numeric)
# =====================================================
titanic.rename(columns={"Gender": "Sex"}, inplace=True)

print(titanic.head())

# =====================================================
# PIE CHART: Male vs Female
# =====================================================
males = (titanic["Sex"] == 1).sum()
females = (titanic["Sex"] == 2).sum()

plt.pie(
    [males, females],
    labels=["Male", "Female"],
    colors=["green", "yellow"],
    explode=(0.15, 0),
    startangle=90,
    autopct="%.2f%%"
)
plt.axis("equal")
plt.show()

# =====================================================
# Survival by Gender
# =====================================================
MaleS = titanic.loc[(titanic["Sex"] == 1) & (titanic["Survived"] == 1)].shape[0]
MaleN = titanic.loc[(titanic["Sex"] == 1) & (titanic["Survived"] == 0)].shape[0]
FemaleS = titanic.loc[(titanic["Sex"] == 2) & (titanic["Survived"] == 1)].shape[0]
FemaleN = titanic.loc[(titanic["Sex"] == 2) & (titanic["Survived"] == 0)].shape[0]

chart = [MaleS, MaleN, FemaleS, FemaleN]
labels = [
    "Survived Male",
    "Not Survived Male",
    "Survived Female",
    "Not Survived Female",
]

plt.pie(
    chart,
    labels=labels,
    explode=[0, 0.05, 0, 0.1],
    startangle=100,
    autopct="%.2f%%"
)
plt.axis("equal")
plt.show()
```
