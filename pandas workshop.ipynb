```python
import matplotlib.pyplot as plt

# Step 1: Create a figure
fig = plt.figure(figsize=(8, 6))

# Step 2: Add subplot in position (2 rows, 2 columns, 2nd plot)
ax2 = fig.add_subplot(2, 2, 2)

# Step 3: Add something to plot (optional example)
ax2.plot([1, 2, 3], [10, 20, 30])
ax2.set_title("Subplot 2")

# Step 4: Show the figure
plt.show()
```

```python
import matplotlib.pyplot as plt

# Step 1: Create a figure
fig = plt.figure(figsize=(8, 6))

# Step 2: Add subplots
ax1 = fig.add_subplot(2, 2, 1)  # subplot 1
ax2 = fig.add_subplot(2, 2, 2)  # subplot 2
ax3 = fig.add_subplot(2, 2, 3)  # subplot 3 (your line)
ax4 = fig.add_subplot(2, 2, 4)  # subplot 4

# Step 3: Add sample plots
ax1.plot([1, 2, 3], [2, 4, 6])
ax2.bar([1, 2, 3], [5, 3, 7])
ax3.scatter([1, 2, 3], [6, 4, 2])
ax4.hist([1, 1, 2, 3, 3, 4, 4])

# Step 4: Add titles for clarity
ax1.set_title("Plot 1")
ax2.set_title("Plot 2")
ax3.set_title("Plot 3")
ax4.set_title("Plot 4")

# Step 5: Display the full figure
plt.tight_layout()
plt.show()
```
