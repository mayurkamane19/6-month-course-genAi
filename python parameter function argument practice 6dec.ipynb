```python
def primes(n):
    out = []
    for num in range(2, n+1):
        if all(num % i != 0 for i in range(2, num)):
            out.append(num)
    return out

print(primes(20))
```

```python
def is_palindrome(n):
    temp, rev = n, 0
    while n > 0:
        rev = rev*10 + n%10
        n //= 10
    return temp == rev

print(is_palindrome(1221))
```

```python
def second_max(lst):
    unique = list(set(lst))
    unique.remove(max(unique))
    return max(unique)

print(second_max([10, 20, 30, 40, 40]))
```

```python
def flatten(lst):
    out = []
    for x in lst:
        if isinstance(x, list):
            out.extend(flatten(x))
        else:
            out.append(x)
    return out

print(flatten([1, [2,3], [4,[5,6]]]))
```

```python
def lcm(a, b):
    greater = max(a, b)
    while True:
        if greater % a == 0 and greater % b == 0:
            return greater
        greater += 1

print(lcm(12, 18))
```

```python
def min_max_scale(x):
    return [(i - min(x)) / (max(x) - min(x)) for i in x]

print(min_max_scale([10, 20, 30, 40]))
```

```python
def standardize(x):
    mean = sum(x)/len(x)
    std = (sum((i-mean)**2 for i in x)/len(x))**0.5
    return [(i-mean)/std for i in x]

print(standardize([1,2,3,4,5]))
```

```python
def accuracy(y_true, y_pred):
    correct = sum(t==p for t,p in zip(y_true, y_pred))
    return correct / len(y_true)

print(accuracy([1,0,1,1], [1,1,1,0]))
```

```python
import math

def softmax(z):
    exp_vals = [math.exp(i) for i in z]
    total = sum(exp_vals)
    return [ev/total for ev in exp_vals]

print(softmax([2.0, 1.0, 0.1]))
```

```python
def gradient_step(m, c, x, y, lr):
    y_pred = m*x + c
    error = y_pred - y
    dm = error * x
    dc = error * 1
    m = m - lr * dm
    c = c - lr * dc
    return m, c

print(gradient_step(1, 0, 4, 10, 0.01))
```

```python
def train_test_split(data, test_size):
    split = int(len(data) * (1 - test_size))
    return data[:split], data[split:]

print(train_test_split([1,2,3,4,5], 0.4))
```

```python
def majority_vote(labels):
    return max(labels, key=labels.count)

print(majority_vote(["A","B","A","A","C"]))
```

```python
def confusion(y_true, y_pred):
    tp = sum(t==1 and p==1 for t,p in zip(y_true,y_pred))
    tn = sum(t==0 and p==0 for t,p in zip(y_true,y_pred))
    fp = sum(t==0 and p==1 for t,p in zip(y_true,y_pred))
    fn = sum(t==1 and p==0 for t,p in zip(y_true,y_pred))
    return tp, tn, fp, fn

print(confusion([1,0,1,0], [1,0,0,1]))
```

```python
def mini_batches(data, batch_size):
    return [data[i:i+batch_size] for i in range(0, len(data), batch_size)]

print(mini_batches([1,2,3,4,5,6,7], 3))
```
